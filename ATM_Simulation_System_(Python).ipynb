```python
#project2

correct_password = "1234"
balance = 1000
max_password_attempts = 3
max_withdraw_attempts = 5

print("---- Welcome to the ATM ----")
input("Insert your card and press Enter to continue...")

password_attempts = 0

while password_attempts < max_password_attempts:
    password = input("Enter your 4-digit PIN: ")

    if password == correct_password:
        print("PIN accepted.\n")
        break
    else:
        password_attempts += 1
        print(f"Incorrect PIN. Attempts left: {max_password_attempts - password_attempts}")

if password_attempts == max_password_attempts:
    print("\nYour card has been blocked due to too many incorrect attempts.")
    exit()

print(f"Your current balance is: ${balance}")

withdraw_attempts = 0

while withdraw_attempts < max_withdraw_attempts:
    try:
        amount = int(input("Enter amount to withdraw: "))
    except ValueError:
        print("Invalid input. Please enter a positive integer.")
        withdraw_attempts += 1
        continue

    # Requirement 1: Must be positive
    if amount <= 0:
        print("Amount must be a positive number.")

    # Requirement 2: Must be multiples of 5
    elif amount % 5 != 0:
        print("Amount must be a multiple of 5 (5, 10, 15, ...).")

    # Requirement 3: Must not exceed balance
    elif amount > balance:
        print("Insufficient funds.")

    else:
        # Valid withdrawal
        balance -= amount
        print(f"\nPlease take your cash: ${amount}")
        print(f"Remaining balance: ${balance}")
        print("Thank you for using our ATM!")
        break

    withdraw_attempts += 1
    print(f"Attempts left: {max_withdraw_attempts - withdraw_attempts}\n")

if withdraw_attempts == max_withdraw_attempts:
    print("\nToo many invalid attempts. Transaction cancelled.")
```
